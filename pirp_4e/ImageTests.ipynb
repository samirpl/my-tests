```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
f = mpimg.imread('../pictures/Samir.tif')
plt.imshow(f,cmap='brg')
plt.colorbar()
```

```python
fcopy = f.copy()
fcopy[   :20,   :  ] = 0 # frame superior
fcopy[-20:  ,   :  ] = 0 # frame inferior
fcopy[   :  ,   :20] = 0 # frame esquerdo
fcopy[   :  ,-20:  ] = 0 # frame direito
plt.imshow(fcopy,cmap='brg')
plt.colorbar()
```

<p style="font-size:30px;"><b>Converting color image to grayscale</b></p>

```python
#Average method
fcopy = f.copy()
fcopy[:] = fcopy.mean(axis=-1,keepdims=1)
plt.imshow(fcopy, cmap='gray')
plt.colorbar()
```

```python
#Luminosity method
fcopy = f.copy()
W = [0.2,0.5,0.3] # weights
W_mean = np.tensordot(fcopy,W, axes=((-1,-1)))[...,None]
fcopy[:] = W_mean.astype(fcopy.dtype)
plt.imshow(fcopy, cmap='gray')
plt.colorbar()
```
